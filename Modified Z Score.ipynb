```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_excel('Movies_revenue.xlsx')
```

```python
df.head()
```

```python
df.revenue.describe()
```

```python
df['revenue_million']= df['revenue'].apply(lambda x:x/1000000)
df.revenue_million.describe()
```

```python
df.head()
```

```python
_,mean,std,*_ =df.revenue_million.describe()
```

```python
mean,std
```

```python
def get_z_score(value,mean,std):
    return (value-mean)/std
```

```python
df['z_score']= df.revenue_million.apply(lambda x: get_z_score(x,mean,std))
df.head()
```

```python
df[df.z_score>3]
```

```python
# Modified Z Score

def get_mad(s):
    median=np.median(s)
    diff=abs(s-median)
    MAD=np.median(diff)
    return MAD
```

```python
MAD = get_mad(df.revenue_million)
median=np.median(df.revenue_million)
MAD, median
```

```python
def get_modified_z_score(x,median,MAD):
    return 0.6745*(x-median)/MAD
```

```python
get_modified_z_score(2787,median,MAD)
```

```python
df['Modified_Z_Score']=df.revenue_million.apply(lambda x:get_modified_z_score(x,median,MAD))
```

```python
df.head()
```

```python
df[df.Modified_Z_Score >3.5]
```
